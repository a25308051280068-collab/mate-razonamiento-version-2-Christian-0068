5 El perímetro del terrenoEnunciado: Un terreno rectangular tiene un largo que mide 5 metros más que el ancho. Si el perímetro es 50 metros, ¿cuánto mide cada lado?Paso 1: Análisis y ResoluciónDefinir las variables:Ancho: $x$Largo: $x + 5$ (porque mide 5 metros más que el ancho).Plantear la ecuación: El perímetro de un rectángulo es la suma de sus cuatro lados ($2 \cdot \text{largo} + 2 \cdot \text{ancho}$).$$2(x + 5) + 2(x) = 50$$Simplificar y despejar:Distribuimos el 2: $2x + 10 + 2x = 50$Sumamos las $x$: $4x + 10 = 50$Pasamos el 10 restando: $4x = 50 - 10 \rightarrow 4x = 40$Dividimos entre 4: $x = \frac{40}{4} = 10$Solución:Ancho: $10$ metros.Largo: $10 + 5 = 15$ metros.

```python
# Problema 5: Dimensiones del terreno
# perimetro_total = 50
# diferencia_largo = 5
perimetro_total=int(input("ingresa el perimetro total:"))
diferencia_largo=int(input("ingresa la diferencia de largo:"))

# La ecuación simplificada fue: 4x + 10 = 50
# Donde 10 es (2 * diferencia_largo)
ancho = (perimetro_total - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"El ancho mide: {int(ancho)} metros")
print(f"El largo mide: {int(largo)} metros")
```
