El problema planteado consiste en desarrollar un programa interactivo en Python, diseñado para nivel principiante y orientado a estudiantes de preparatoria, que permita calcular propiedades básicas de figuras tridimensionales y bidimensionales: el área y el perímetro de un círculo, y el volumen de una esfera.

* El programa se estructura de manera procedural (sin programación orientada a objetos) y se mantiene en un único archivo para facilitar su comprensión y ejecución.
* Se implementa un ciclo while True que mantiene el programa activo, mostrando un menú interactivo con emojis para mejorar la experiencia visual.
* Las decisiones se manejan con if, elif y else, los cuales dirigen el flujo hacia la función correspondiente: areacirculo(), perimetrocirculo() o volumenesfera().
* Cada función solicita el radio únicamente cuando es invocada, realiza el cálculo utilizando la librería estándar math y muestra el resultado formateado. No se incluye validación de datos de entrada, priorizando la simplicidad y el enfoque en estructuras de control básicas.
* La opción 4 permite salir del ciclo y finalizar el programa de manera ordenada.


```python
# ============================================================================
#  PROGRAMA DE CÁLCULO GEOMÉTRICO 📐
# Elaborao por Christian Garcia
# Cálculo de área y perímetro de círculo, y volumen de esfera
# Nivel: Principiante - Preparatoria
# ============================================================================

import math

# ============================================================================
# FUNCIÓN: areacirculo()
# Calcula el área de un círculo
# Fórmula: A = π × r²
# ============================================================================
def areacirculo():
    """🔵 Calcula el área de un círculo"""
    print("\n" + "="*50)
    print("🔵   CÁLCULO DEL ÁREA DEL CÍRCULO   🔵")
    print("="*50)

    # Solicitar el radio
    radio = float(input("📏 Ingresa el radio del círculo: "))

    # Calcular el área: A = π × r²
    area = math.pi * (radio ** 2)

    # Mostrar resultado
    print("\n✨✨✨ RESULTADO ✨✨✨")
    print(f"📐 Radio: {radio}")
    print(f"🔵 Área del círculo: {area:.4f}")
    print(f"🔵 Área del círculo (redondeada): {round(area, 2)}")
    print("="*50 + "\n")

# ============================================================================
# FUNCIÓN: perimetrocirculo()
# Calcula el perímetro (circunferencia) de un círculo
# Fórmula: P = 2 × π × r
# ============================================================================
def perimetrocirculo():
    """🔴 Calcula el perímetro de un círculo"""
    print("\n" + "="*50)
    print("🔴   CÁLCULO DEL PERÍMETRO DEL CÍRCULO   🔴")
    print("="*50)

    # Solicitar el radio
    radio = float(input("📏 Ingresa el radio del círculo: "))

    # Calcular el perímetro: P = 2 × π × r
    perimetro = 2 * math.pi * radio

    # Mostrar resultado
    print("\n✨✨✨ RESULTADO ✨✨✨")
    print(f"📐 Radio: {radio}")
    print(f"🔴 Perímetro del círculo: {perimetro:.4f}")
    print(f"🔴 Perímetro del círculo (redondeada): {round(perimetro, 2)}")
    print("="*50 + "\n")

# ============================================================================
# FUNCIÓN: volumenesfera()
# Calcula el volumen de una esfera
# Fórmula: V = (4/3) × π × r³
# ============================================================================
def volumenesfera():
    """⚽ Calcula el volumen de una esfera"""
    print("\n" + "="*50)
    print("⚽   CÁLCULO DEL VOLUMEN DE LA ESFERA   ⚽")
    print("="*50)

    # Solicitar el radio
    radio = float(input("📏 Ingresa el radio de la esfera: "))

    # Calcular el volumen: V = (4/3) × π × r³
    volumen = (4/3) * math.pi * (radio ** 3)

    # Mostrar resultado
    print("\n✨✨✨ RESULTADO ✨✨✨")
    print(f"📐 Radio: {radio}")
    print(f"⚽ Volumen de la esfera: {volumen:.4f}")
    print(f"⚽ Volumen de la esfera (redondeada): {round(volumen, 2)}")
    print("="*50 + "\n")

# ============================================================================
# FUNCIÓN: menu()
# Muestra el menú principal y controla la ejecución del programa
# ============================================================================
def menu():
    """📋 Muestra el menú y controla el programa"""
    opcion = 0

    while opcion != 4:
        # Mostrar menú elegante con emojis
        print("\n" + "🌟"*30)
        print("🌟" + " "*28 + "🌟")
        print("🌟   📐 MENÚ PRINCIPAL 📐   ")
        print("" + " "*28 + "🌟")
        print("🌟"*30)
        print("\n   🔹 1. Calcular área del círculo")
        print("   🔹 2. Calcular perímetro del círculo")
        print("   🔹 3. Calcular volumen de la esfera")
        print("   🔹 4. Salir del programa")
        print("\n" + "─"*50)

        # Leer opción del usuario
        opcion = int(input("💡 Selecciona una opción (1-4): "))

        # Usar if, elif, else para procesar la opción
        if opcion == 1:
            # Llamar función para calcular área
            areacirculo()
        elif opcion == 2:
            # Llamar función para calcular perímetro
            perimetrocirculo()
        elif opcion == 3:
            # Llamar función para calcular volumen
            volumenesfera()
        elif opcion == 4:
            # Opción de salida
            print("\n" + "🎉"*30)
            print("🎉" + " "*28 + "🎉")
            print("🎉   ¡GRACIAS POR USAR EL PROGRAMA!   🎉")
            print("🎉" + " "*28 + "🎉")
            print("🎉"*30)
            print("\n   👋 ¡Hasta pronto! 👋\n")
        else:
            # Opción no válida
            print("\n" + "⚠️"*25)
            print("⚠️   OPCIÓN NO VÁLIDA   ⚠️")
            print("⚠️   Por favor, elige 1, 2, 3 o 4   ⚠️")
            print("⚠️"*25 + "\n")

# ============================================================================
#  INICIO DEL PROGRAMA PRINCIPAL
# ============================================================================
if __name__ == "__main__":
    # Llamar a la función del menú para iniciar el programa
    print("\n" + "🎓"*30)
    print("🎓" + " "*28 + "🎓")
    print("🎓   BIENVENIDO AL PROGRAMA GEOMÉTRICO   🎓")
    print("🎓" + " "*28 + "🎓")
    print("🎓"*30)
    print("\n   📚 Cálculo de figuras geométricas 📚")
    print("   🎯 Nivel: Preparatoria 🎯\n")

    # Iniciar el menú
    menu()
```
